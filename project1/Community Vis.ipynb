```python
import networkx as nx
import community

import matplotlib.pyplot as plt
%matplotlib inline

G_fb = nx.read_edgelist("facebook_combined.csv", delimiter=',', create_using = nx.Graph())
parts = community.best_partition(G_fb)
values = [parts.get(node) for node in G_fb.nodes()]

#Create network layout for visualizations
spring_pos = nx.spring_layout(G_fb)

plt.axis("off")
nx.draw(G_fb, pos = spring_pos, cmap = plt.get_cmap("jet"), node_color = values, node_size = 35, with_labels = False, edgecolors='black')
plt.show()
```
